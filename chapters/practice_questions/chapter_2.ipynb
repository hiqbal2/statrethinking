###### (1) WWW

```python
n = 3
k = 3
prior = np.linspace(0,1, 1000) # same probability for all fractions of water
likelihood = stats.binom.pmf(k, n, prior)
unstd_posterior = likelihood * prior
posterior = unstd_posterior / unstd_posterior.sum()

plt.plot(prior, posterior, 'o-', label='success = {}\ntosses = {}'.format(k, n))
plt.xlabel('probability of water')
plt.ylabel('posterior probability')
plt.legend(loc=0);
```

$\text{P( Mars )} = 0.5$

$$
\begin{align}
\text{P(Earth | land)} &= \frac{\text{P(land | Earth) P(Earth)}}{\text{P(land)}} \\
& = \frac{\text{P(land | Earth) P(Earth)}}{\text{(land | Earth) P(Earth) + P(land | Mars) P(Mars)}} \\
& = \frac{(1 - 0.7) * 0.5}{(1 - 0.7) * 0.5 + (1 - 0) * 0.5} \\
& = \frac{0.3 * 0.5}{0.3 * 0.5 + 0.5} \\
& = 0.23
\end{align}
$$

##### 2M4

P(card 1) = P(card 2) = P(card 3) = 1/3  
P(white | card 1) = 0  
P(black | card 1) = 1  
P(white | card 2) = 0.5  
P(black | card 2) = 0.5  
P(white | card 3) = 1  
P(black | card 3) = 0  

$$
\begin{align}
\text{P(one side black)} &= \text{P(black | card 1) P(card 1) + P(black | card 2) P(card 2) + P(black | card 3) P(card 3)} \\
& = 1 * \frac{1}{3} + 0.5 * \frac{1}{3} + 0 * \frac{1}{3} \\
& = 0.5
\end{align}
$$

Therefore,
$$
\begin{align}
\text{P(second side black | one side black)} &= \frac{\text{P(both sides black)}}{\text{P(one side black)}} \\
& = \frac{1/3}{1/2} \\
& = \frac{2}{3}
\end{align}
$$
