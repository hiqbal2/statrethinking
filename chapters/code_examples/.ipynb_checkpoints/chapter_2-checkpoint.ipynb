# Small Worlds and Large Worlds

```python
%matplotlib inline
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd
import pymc3 as pm
```

## The Garden of Forking Data

Reproducing the example of updating plausibility of getting $D_{new}$ using the formula:

$$
\text{plausibility of} \textit{ p } \text{after D}_{new} = \frac{\text{ways p can produce D}_{new} \times \text{prior plausibility p}}{\text{sum of products}}
$$

```python
prior = np.array([1/5] * 5) # assuming each composition of the bag of marbles is equally possible
ways = np.array([0, 3, 8, 9, 0]) # ways in which each composition can generate sequence of black, white and black

sum_of_products = np.sum(prior * ways)
# updated plausibility of of each composition
updated_plausibility = (prior * ways) / sum_of_products

print(updated_plausibility)
```

## Building a model

If we always assume that each fraction of water level has the same plausibility, then we can simulate the updated plausibility as we keep observing newer observations. In the book, the observations for water and land were as follows: W, L, W, W, W, L, W, L, W. The updated plausibility as we observe more, as compared to a uniform (equal) plausibility is as follows:

```python
# Water = 1, Land = 0
observed_seq = np.array([1,0,1,1,1,0,1,0,1])
```

```python
# Make multiple subplots for all 9 observations
fig, axs = plt.subplots(3, 3, figsize=(12, 12), sharey='row',)

# assuming different outcomes of fraction of earth being water
p_grid = np.linspace(0,1,101)

# NOTE: always assuming all outcomes equally likely
prior = np.ones(101) / np.sum(np.ones(101))

# We also compare the pluasibility distribution to when we have less data. 
# We start with having no data and assuming the previous updated plausibility was equal for all water fractions
previous_outcome = prior

# As we get more observations
for ax, i in zip(axs.flatten(), range(1,len(observed_seq) + 1)):
    current_observations = observed_seq[:i]
    
    # Total observations (i.e., number of tosses)
    n = len(current_observations)
    
    # Number of times water lands on index finger
    k = np.sum(current_observations)
    
    # ways to see observed data (k water out of n tosses) given fraction (p_grid) of earth being water
    # this comes from the binomial distribution formula
    likelihood = stats.binom.pmf(k=k, n=n, p=p_grid)
    
    # updated probability after observing new toss
    ways = likelihood * prior
    posterior = ways / np.sum(ways)
    
    # plot the previos posterior with 1 less data point and the new posterior
    ax.plot(p_grid, previous_outcome, 'b--', p_grid, posterior, 'b')  
    previous_outcome = posterior
```

##### Prior

Replicating Figure 2.6

```python
p_grid = np.linspace(0,1,101)
n = 9
k = 6
```

```python
flat_prior = np.ones(101) / np.sum(np.ones(101))
step_prior = np.where(p_grid > 0.5, 1, 0)
peaked_prior = np.exp(-np.abs(p_grid - 0.5)*5) # using a laplace distribution pdf
```

```python
# Make multiple subplots for all 9 plots
fig, axs = plt.subplots(3, 3, figsize=(12, 12))
axes = axs.flatten()

priors = [flat_prior, step_prior, peaked_prior]
likelihood = stats.binom.pmf(k, n, p_grid)
for i in range(3):
    prior = priors[i]
    ways = likelihood * prior
    posterior = ways / np.sum(ways)
    
    idx = 3 * i
    axes[idx].plot(p_grid, prior)
    axes[idx+1].plot(p_grid, likelihood)
    axes[idx+2].plot(p_grid, posterior)
    
```

##### Grid Approximation

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axes = axs.flatten()

for idx, val in enumerate([5, 20]):
    
    # Define grid
    p_grid = np.linspace(0, 1, val)

    # define prior
    prior = np.array([1,] * val)

    # Compute likelihood at each value in p_grid
    likelihood = stats.binom.pmf(k=6, n=9, p=p_grid)

    # compute product of likelihood and prior
    unstd_posterior = likelihood * prior

    # standardize the posterior, so fit sums to 1
    posterior = unstd_posterior / np.sum(unstd_posterior)
    
    axes[idx].plot(p_grid, posterior, marker='o')
    axes[idx].set_ylabel('posterior probability')
    axes[idx].set_xlabel('probability of water')
    axes[idx].set_title('{val} points'.format(val=val))
```

##### Quadratic Approximation

Copying solution from https://github.com/pymc-devs/resources/blob/master/Rethinking/Chp_02.ipynb

```python
# Water = 1, Land = 0
data = np.array([1,0,1,1,1,0,1,0,1])
```

```python
# uses 'L-BFGS-B' for optimisation
with pm.Model() as normal_aproximation:
    p = pm.Uniform('p', 0, 1)
    w = pm.Binomial('w', n=len(data), p=p, observed=data.sum())
    mean_q = pm.find_MAP()
    std_q = ((1/pm.find_hessian(mean_q, vars=[p]))**0.5)[0]
```

```python
mean_q['p'], std_q
```

```python
norm = stats.norm(mean_q, std_q)
prob = .89
z = stats.norm.ppf([(1-prob)/2, (1+prob)/2])
pi = mean_q['p'] + std_q * z 
pi
```

```python
fig, axs = plt.subplots(1, 3, figsize=(12, 6))
axes = axs.flatten()

for i, n in enumerate([9,18,36]):
    k = n*(2/3)
    with pm.Model() as normal_aproximation:
        p = pm.Uniform('p', 0, 1)
        w = pm.Binomial('w', n=n, p=p, observed=k)
        mean_q = pm.find_MAP()
        std_q = ((1/pm.find_hessian(mean_q, vars=[p]))**0.5)[0]
    
    # analytical calculation

    x = np.linspace(0, 1, 100)
    plt.plot()

    # quadratic approximation
    axes[i].plot(x, stats.norm.pdf(x, mean_q['p'], std_q), label='Quadratic approximation')
    axes[i].plot(x, stats.beta.pdf(x , k+1, n-k+1), label='True posterior')
    
    axes[i].set_ylabel('Density', fontsize=14)
    axes[i].set_xlabel('Proportion water', fontsize=14)
    axes[i].set_title('n = {}'.format(n), fontsize=14)
```
